```python
X=data.drop('FWI',axis=1)
y=data['FWI']
```

```python
def correlation(data,threshold):
    corr_matrix=data.corr()
    corr_features=set()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]>threshold):
                corr_features.add(corr_matrix.columns[i])
    return corr_features
```

```python
corr_features=correlation(X,0.85)
```

```python
corr_features
```

```python
X.drop(corr_features,axis=1,inplace=True)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_train.shape,y_train.shape
```

## Standarization 

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```python
Xtrain_t=scalar.fit_transform(X_train)
```

```python
Xtest_t=scalar.transform(X_test)
```

```python
X_train_new=pd.DataFrame(Xtrain_t,columns=X_train.columns)
```

```python
X_test_new=pd.DataFrame(Xtest_t,columns=X_test.columns)
```

```python
X_test_new.head()
```

```python
X_test.head()
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,5))
sns.boxplot(data=X_train,ax=axes[0])
axes[0].set_title('X_train before Scaling')
sns.boxplot(data=X_train_new,ax=axes[1])
axes[1].set_title('X_train after Scaling')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
```

```python
linreg.fit(X_train,y_train)
```

```python
y_pred_train=linreg.predict(X_train)
```

```python
y_pred=linreg.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
r_score=r2_score(y_test,y_pred)
```

```python
print(np.sqrt(mse))
print(mse)
print(r_score)
plt.scatter(y_test,y_pred)
```

```python
mse2=mean_squared_error(y_train,y_pred_train)
r_score2=r2_score(y_train,y_pred_train)
```

```python
print(np.sqrt(mse2))
print(mse2)
print(r_score2)
plt.scatter(y_train,y_pred_train)
```

## Lasso

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
```

```python
lasso.fit(X_train,y_train)
```

```python
lasso_pred_train=lasso.predict(X_train)
lasso_pred_test=lasso.predict(X_test)
```

```python
mse2=mean_squared_error(y_train,lasso_pred_train)
r_score2=r2_score(y_train,lasso_pred_train)
print(np.sqrt(mse2))
print(mse2)
print(r_score2)
plt.scatter(y_train,lasso_pred_train)
```

```python
mse2=mean_squared_error(y_test,lasso_pred_test)
r_score2=r2_score(y_test,lasso_pred_test)
print(np.sqrt(mse2))
print(mse2)
print(r_score2)
plt.scatter(y_test,lasso_pred_test)
```

```python
from sklearn.linear_model import LassoCV
cv=LassoCV(cv=5)
```

```python
cv.fit(X_train,y_train)
```

```python
cv_pred_test=cv.predict(X_test)
mse3=mean_squared_error(y_test,cv_pred_test)
r_score3=r2_score(y_test,cv_pred_test)
print(np.sqrt(mse3))
print(mse3)
print(r_score3)
plt.scatter(y_test,cv_pred_test)
```
